# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
QUERY="""
SELECT Respondent, ConvertedComp
FROM Master
"""
df_ConvertedComp=pd.read_sql_query(QUERY, conn)
df_ConvertedComp['ConvertedComp'].plot(kind='hist', figsize=(10,7))
plt.title('Converted Compensation Histogram')
plt.ylabel('Respondents')
plt.xlabel('Converted Compensation Exchange Rate')
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
QUERY="""
SELECT Respondent, Age 
FROM Master
"""
df_Age=pd.read_sql_query(QUERY, conn)
df_Age['Age'].plot(kind='box', figsize=(10,7))
plt.title('Age of the Respondents')
plt.ylabel('Age Interval')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
QUERY="""
SELECT Age, WorkWeekHrs
FROM Master
"""
df_Age=pd.read_sql_query(QUERY, conn)
df_Age.plot(kind='scatter', x='Age', y='WorkWeekHrs', figsize=(10,6), color='darkgreen')
plt.title('Age Correlation to WorkWeekHrs amongst Respondents')
plt.ylabel('WorkWeekHrs')
plt.xlabel('Age')
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
QUERY="""
SELECT CodeRevHrs, WorkWeekHrs, Age
FROM Master
"""
df_bubble=pd.read_sql_query(QUERY, conn)
norm_Age=(df_bubble['Age']-df_bubble['Age'].min()) / (df_bubble['Age'].max()-df_bubble['Age'].min())
```

```python
df_bubble.plot(kind='scatter',
                   x='WorkWeekHrs',
                   y='CodeRevHrs',
                   figsize=(14,8),
                   alpha=0.5,
                   color='orange',
                   s=norm_Age*2000+10
                  )
plt.title('WorkWeekHrs vs CodeRevHrs of Respondents')
plt.ylabel('CodeRevHrs')
plt.xlabel('WorkWeekHrs')
plt.legend(['Age'], loc='upper right', fontsize='large')
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
QUERY="""
SELECT * 
from DatabaseDesireNextYear
"""
df_DbDesire=pd.read_sql_query(QUERY, conn)
df_DbGroup=df_DbDesire.groupby('DatabaseDesireNextYear', axis=0).count()
df_DbGroup.sort_values(['Respondent'], ascending=False, inplace=True)
df_pie=df_DbGroup.head()
colors_list=['gold', 'yellowgreen', 'lightcoral', 'lightblue', 'pink']
df_pie['Respondent'].plot(kind='pie',
                   figsize=(15,6),
                   autopct='%1.1f%%',
                   startangle=90,
                   shadow=True,
                   colors=colors_list,
                  )
plt.title('Top 5 Most Desired Databases to Learn Next Year')
plt.axis('equal')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY="""
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM Master
WHERE Age between 30 and 35
"""
df_stack=pd.read_sql_query(QUERY, conn)

print(df_stack.groupby('Age').median())

order=['WorkWeekHrs', 'CodeRevHrs']
df_stack.groupby('Age').median().plot.bar(stacked=True)
plt.ylim(ymin=0)
plt.title('Median hours by Age')
plt.ylabel('Median Hours')
plt.xlabel('Age')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY="""
SELECT ConvertedComp, Age
FROM Master
"""

df_line=pd.read_sql_query(QUERY, conn)
df_line=df_line.loc[(df_line['Age']>=45) & (df_line['Age']<=60)].reset_index().drop(columns=['index'])
df_line.dropna(inplace=True)

df_line=df_line.groupby('Age', axis=0).median()

df_line.plot(kind='line')
plt.title('Median Converted Comp for Ages from 45 to 60')
plt.xlabel('Age')
plt.ylabel('Median Converted Comp')
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
QUERY="""
SELECT MainBranch, count(*)
FROM Master
GROUP BY MainBranch
"""
df_bar=pd.read_sql_query(QUERY, conn)
df_bar.plot(kind='barh', x='MainBranch', color='purple')
plt.title('The Main Branch of Respondents')
df_bar
```
